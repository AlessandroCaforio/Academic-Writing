### Understanding `np.random.default_rng`

The `default_rng(seed)` function initializes a `Generator` object. Once you have this object (often named `rng`), you call methods on it to produce different distributions.

```python
# Initialize the generator
example_rng = np.random.default_rng(seed=123)

# Generate different types of data using the same generator object
rand_floats = example_rng.random(3)          # 3 floats between 0 and 1
rand_ints = example_rng.integers(0, 10, 5)    # 5 integers between 0 and 9
rand_normal = example_rng.standard_normal(3) # 3 numbers from standard normal distribution

print(f"Floats: {rand_floats}")
print(f"Integers: {rand_ints}")
print(f"Normal Dist: {rand_normal}")
```
